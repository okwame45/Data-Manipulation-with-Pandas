### Exercises: For Loops

Practice using for loops with these exercises.

**Exercise 1:** Given a list of weekly yields: [45, 52, 48, 55, 60, 58, 52], calculate:
- Total yield
- Average yield
- Number of weeks above 50 tons

````python
# Your solution here
weekly_yields = [45, 52, 48, 55, 60, 58, 52]

total = 0
above_50 = 0
for yields in weekly_yields:
    total = sum(weekly_yields)
    average_yield = total/len(weekly_yields)
    
    if yields > 50:
       above_50 += 1

print("Total yield: ",total)
print("Average yield: ",average_yield)
print("Number of weeks above 50 tons: ",above_50)
````

**Exercise 2:** Loop through this list of temperatures and count how many days were:
- Cold (< 20°C)
- Moderate (20-30°C)
- Hot (> 30°C)

Temperatures: [18, 22, 25, 32, 28, 19, 31, 26, 24, 33]

````python
# Your solution here
temperatures = [18, 22, 25, 32, 28, 19, 31, 26, 24, 33]

cold_days = 0
moderate_days = 0
hot_days = 0

for temp in temperatures:
    if temp < 20:
        cold_days +=1
    elif temp <= 30 and temp >= 20:
        moderate_days += 1
    elif  temp > 30:
        hot_days += 1

print("Cold days: ",cold_days)
print("Moderate days: ",moderate_days)
print("Hot days: ",hot_days)
````

### Exercises: While Loops

Practice using while loops with these exercises.

**Exercise 1:** Write a program that simulates adding water to a tank.
- Start with 100 liters
- Target: 500 liters
- Add 50 liters per minute
- Print the water level each minute
- Show total time when target is reached

````python
# Your solution here
current_water = 100  # liters
target_water = 500  # liters
fill_rate = 50  # liters per minute
minutes = 0

while current_water != target_water:
    current_water += fill_rate
    minutes += 1
    print(f"water level is {current_water} after {minutes} minutes")

print(minutes)
````

**Exercise 2:** Simulate a savings program for a farmer.
- Starting savings: ₦50,000
- Goal: ₦500,000
- Monthly saving: ₦35,000
- Count how many months it takes to reach the goal

````python
# Your solution here
current_savings = 50000  # Naira
goal = 500000  # Naira
monthly_saving = 35000  # Naira
months = 0

print("Farmer Savings Program")
print("="*40)
print(f"Starting savings: ₦{current_savings:,}")
print(f"Goal: ₦{goal:,}")
print(f"Monthly saving: ₦{monthly_saving:,}\n")

while current_savings < goal:
    current_savings += monthly_saving
    months += 1

print(f"number of months it takes to reach the goal is {months}")
````
